```python
import pandas as pd
data_pret = pd.read_csv(r"C:\Users\DELL E5580\Desktop\tps_data\prets.csv")
data_pret
```

```python
# ajouter une colonne "taux_de_remboursement" et arrondir a 2 chiffres apres la virgule

data_pret["taux_de_remboursement"] = round((data_pret["remboursement"]*100) / data_pret["revenu"], 2)
data_pret
```

```python
# renommer la colonne taux par taux_d_interet

data_pret.rename(columns = {"taux":"taux_d_interet"}, inplace = True)
data_pret
```

```python
# Calculer le cout total

data_pret["cout_total"] = data_pret["remboursement"] * data_pret["duree"]
data_pret
```

```python
data_pret["benefice"] = round((data_pret["cout_total"] * data_pret["taux_d_interet"]) / 2400, 2)
data_pret
```

```python
# Calculer le nombre de personnes ayant dépassé le taux d’endettement autorisé 
nombre_depassement = (data_pret['remboursement'] / data_pret['revenu']) > 0.35
nombre_personnes_depassement = nombre_depassement.sum()
print("Nombre de personnes ayant dépassé le taux d'endettement autorisé :", nombre_personnes_depassement)
```

```python
# Ajouter la variable risque
data_pret['risque'] = ['oui' if taux_depassement else 'non' for taux_depassement in nombre_depassement]
data_pret
```

```python
# Nombre de prêts automobiles accordés et coût total moyen

nombre_prets_auto = data_pret[data_pret['type'] == 'automobile'].shape[0]
cout_total_moyen_auto = data_pret[data_pret['type'] == 'automobile']['cout_total'].mean()
print("Nombre de prêts automobiles accordés :", nombre_prets_auto)
print("Coût total moyen des prêts automobiles :", cout_total_moyen_auto)
```

```python
# Bénéfice total dégagé par chaque ville, par type de prêt 
benefice_par_ville_type_pret = data_pret.groupby(['ville', 'type'])['benefice'].sum()
print(benefice_par_ville_type_pret)
```

```python
# Bénéfices moyens réalisés par chaque ville, pour chaque type de prêt, sous forme de tableau à double entrée 
benefices_moyens = data_pret.pivot_table(index='ville', columns='type', values='benefice', aggfunc='mean')
print(benefices_moyens)
```

```python
# Ville la plus intéressante pour développer les prêts immobiliers
ville_interessante_immo = benefice_par_ville_type_pret.loc[:, 'immobilier'].idxmax()
print("Ville la plus intéressante pour développer les prêts immobiliers :", ville_interessante_immo)
```
