```python
import matplotlib.pyplot as plt
import numpy as np

# creation d'un graphique lineaire
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y)
# plt.show
```

```python
# Transformation du graphique en nuage de point
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, "o")
```

```python
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, "x")
```

```python
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, marker = "o")
```

```python
# mise en forme du graphe
plt.plot(x, y, marker = "o", lw = 10, ms = 20, mec = "r", mfc = "g" )
plt.show
```

```python
# Ajouter un titre au graphe
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, marker = "o")
plt.title("Graphique lineaire", fontweight = "bold", loc = "right", color = "red")
plt.show
```

```python
# Ajouter un titre aux differents axes du graphe
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, marker = "o")
plt.xlabel("Axe des X")
plt.ylabel("Axe des Y")
plt.show
```

```python
# Ajouter d'ajouter le quadrillage 
x = np.array([1, 4, 6])
y = np.array([2, 5, 5])
plt.plot(x, y, marker = "o")
plt.grid()
```

```python
# Faire les nuages de point
x = np.random.random(100)
y = np.random.random(100)
plt.scatter(x, y, marker = "*")
plt.show
```

```python
# Faire les nuages de point / Modifier les tailles des point
x = np.random.random(10)
y = np.random.random(10)
sizes = np.array([22, 56, 19, 57, 29, 20, 22, 36, 12, 9])
colors = np.random.randint(50, size = (10))
plt.scatter(x, y, s = sizes, c = colors, cmap = "Dark2")
plt.show
```
