```python
#Gestion des stocks

liste = [["Yaourt", 20, 450], ["Pain", 50, 550], ["Sucre", 10, 700]]

def ajouter(gestion):
    nom = input("Entrez le nom du produit : ")
    quantite = int(input("Entrez la quantité du produit : "))
    pu = int(input("Entrez le prix unitaire du produit : "))
    liste1 = [nom, quantite, pu]
    gestion.append(liste1)
    print(gestion)

def rechercher(gestion):
    nom = input("Entrez le nom du produit : ")
    for i in gestion:
        if nom == i[0]:
            print("Quantité =", i[1], "PU =", i[2])
            return
    print("Le produit n'est pas disponible")

def mettre_a_jour(gestion):
    nom = input("Entrez le nom du produit : ")
    for i in gestion:
        if nom == i[0]:
            PU = int(input("Entrez le nouveau prix unitaire du produit : "))
            i[2] = PU
            print(gestion)
            return
    print("Le produit n'est pas disponible")

def afficher(gestion):
    cout_total = 0
    for i in gestion:
        cout_total += i[1] * i[2]
    print("Coût total des produits en stock :", cout_total)

def stock(gestion):
    while True:
        print("MENU")
        print("1. Ajouter")
        print("2. Rechercher")
        print("3. Mettre à jour")
        print("4. Afficher")
        print("5. Quitter")

        choix = int(input("Faites votre choix : "))

        if choix == 1:
            ajouter(gestion)
        elif choix == 2:
            rechercher(gestion)
        elif choix == 3:
            mettre_a_jour(gestion)
        elif choix == 4:
            afficher(gestion)
        elif choix == 5:
            print("A bientôt")
            break
        else:
            print("Choix incorrect")

stock(liste)
```
