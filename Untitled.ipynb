```python
import pandas as pd
data_pret = pd.read_csv(r"C:\Users\DELL E5580\Desktop\tps_data\prets.csv")
data_pret
```

```python
# ajouter une colonne "taux_de_remboursement" et arrondir a 2 chiffres apres la virgule

data_pret["taux_de_remboursement"] = round((data_pret["remboursement"]*100) / data_pret["revenu"], 2)
data_pret
```

```python
# renommer la colonne taux par taux_d_interet

data_pret.rename(columns = {"taux":"taux_d_interet"}, inplace = True)
data_pret
```

```python
# Calculer le cout total

data_pret["cout_total"] = data_pret["remboursement"] * data_pret["duree"]
data_pret
```

```python
data_pret["benefice"] = round((data_pret["cout_total"] * data_pret["taux_d_interet"]) / 2400, 2)
data_pret
```
