# gestion de stock dictionnaire

```python
# generer 10 produits 

import random
import string

# fonction generate product
# (_) variable de parcours qui ne sera pas utilise dans le programme

def generate_product():
    produits = {}
    for _ in range(5):
        nom = ''.join(random.choices(string.ascii_letters, k = random.randint(5,10)))
        quantite = random.randint(5,10)
        prix_unitaire = random.randint(2000, 5000)
        produits[nom] = {"quantite" : quantite, "prix_unitaire": prix_unitaire}
        
    return produits
produits = generate_product()
produits

# gestion des stocks

# fonction d'ajout

def add_product():
    nom = input("Entrez le nom du produit : ")
    quantite = int(input("Entrez la quantité du produit : "))
    prix_unitaire = int(input("Entrez le prix unitaire du produit : "))
    new_product = {"quantite" : quantite, "prix_unitaire": prix_unitaire}
    produits[nom]= new_product
    print("produit ajoute avec succes", produits)
    return produits

# fonction rechercher un produit
def search_product():
    nom = input("Entrez le nom du produit : ")
    produit = produits.get(nom)
    if produit:
        print("nom : {}".format(nom))
        print("Quantite : {}".format({produit['quantite']})) 
        print(f"prix_unitaire : {produit['prix_unitaire']}")
    else:
        print("Le produit n'est pas disponible")

# fonction pour mettre_a_jour produit
def update_product():
    nom = input("Entrez le nom du produit : ")
    produit = produits.get(nom)
    if produit:
        new_price=int(input("Entrez le nouveau prix"))
        produit['prix_unitaire'] = new_price
        print("Le produit a ete mise a jour avec succes")
        print(produits)
    else:
        print("Le produit est non disponible dans le stock")

# fonction pour afficher le cout total des produits

def Cout_total_product():
    Cout_total = sum(produit['quantite'] * produit['prix_unitaire'] for produit in produits.values())
    print ("le cout total des produits est: ", Cout_total)
    
# Menu

while True:
        print("MENU")
        print("1 -> Ajouter")
        print("2 -> Rechercher")
        print("3 -> Mettre à jour")
        print("4 -> Afficher")
        print("5 -> Quitter")

        choix = int(input("Faites votre choix : "))

        if choix == 1:
            liste_produit = add_product()
            print (liste_produit)
        
        elif choix == 2:
            #liste_produit = 
            search_product()
           # print (liste_produit)
            
            #print ("Le produit n'est pas disponible")
        elif choix == 3:
            update_product()
            
        elif choix == 4:
            Cout_total_product()
            
        elif choix == 5:
            print("A bientôt")
            break
        else:
            print("Choix incorrect, entrez un nombre entre 1 et 5")
#add_product(produits)
```
